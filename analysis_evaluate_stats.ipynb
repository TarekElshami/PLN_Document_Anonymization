```python
import os
import subprocess
import re
import pandas as pd
import matplotlib.pyplot as plt
import shutil
import tempfile

# Configuraci√≥n
RESULTS_DIR = "results"
SYSTEM_DIRS = [
    "procesados_meddocan/systemLlama3.3/entidades/",
    "procesados_meddocan/systemLlama3.3/etiquetado/",
    "procesados_meddocan/systemPhi4/entidades/",
    "procesados_meddocan/systemPhi4/etiquetado/"
]
SYSTEM_WRONG = "systemWrong"
EVAL_TYPES = ["ner", "spans"]
GRAPH_FILE = os.path.join(RESULTS_DIR, "metrics__{}__{}.png")

def run_evaluation(test_dir, system_dir, eval_type):
    cmd = ["python", "evaluate.py", "i2b2", eval_type, test_dir, system_dir]
    result = subprocess.run(cmd, capture_output=True, text=True)
    return result.stdout

def sort_by_prompt_number(lst):
    def extract_number(s):
        match = re.search(r'(\d+)', s)
        return int(match.group(1)) if match else float('inf')
    return sorted(lst, key=extract_number)

def sanitize_filename(name):
    return name.strip("/").replace("/", "_")

def parse_evaluation_output(output):
    stats = {"Leak": 0, "Precision": 0, "Recall": 0, "F1": 0}
    for line in output.splitlines():
        if "Leak" in line and not "Leakage" in line:
            leak = re.search(r"Leak\s+([\d.]+)", line)
            if leak: stats["Leak"] = float(leak.group(1))
        elif "Precision" in line:
            prec = re.search(r"Precision\s+([\d.]+)", line)
            if prec: stats["Precision"] = float(prec.group(1))
        elif "Recall" in line:
            rec = re.search(r"Recall\s+([\d.]+)", line)
            if rec: stats["Recall"] = float(rec.group(1))
        elif "F1" in line:
            f1 = re.search(r"F1\s+([\d.]+)", line)
            if f1: stats["F1"] = float(f1.group(1))
    return stats

def generar_grafica_unica(df, system_name, eval_type):
    plt.figure(figsize=(14, 7))
    if eval_type != "spans":
        plt.plot(df["Config"], df["Leak"], label="Leak", marker="o", linestyle="--", color="red")
    plt.plot(df["Config"], df["Precision"], label="Precision", marker="s", color="blue")
    plt.plot(df["Config"], df["Recall"], label="Recall", marker="^", color="green")
    plt.plot(df["Config"], df["F1"], label="F1", marker="D", color="purple")

    plt.title(f"M√©tricas ({eval_type}) - {system_name}", fontsize=14)
    plt.xlabel("Configuraci√≥n", fontsize=12)
    plt.ylabel("Score", fontsize=12)
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()

    graph_path = GRAPH_FILE.format(system_name, eval_type)
    plt.savefig(graph_path)
    print(f"‚úÖ Gr√°fica guardada en: {graph_path}")
    plt.show()

def generar_grafica_comparativa(df_ner, df_spans, system_name):
    plt.figure(figsize=(14, 7))

    for metric in ["Precision", "Recall", "F1"]:
        plt.plot(df_ner["Config"], df_ner[metric], label=f"{metric} (NER)", linestyle="--", marker="o")
        plt.plot(df_spans["Config"], df_spans[metric], label=f"{metric} (Spans)", linestyle="-", marker="s")

    plt.title(f"Comparativa NER vs Spans - {system_name}", fontsize=14)
    plt.xlabel("Configuraci√≥n", fontsize=12)
    plt.ylabel("Score", fontsize=12)
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()

    graph_path = GRAPH_FILE.format(system_name, "comparativa")
    plt.savefig(graph_path)
    print(f"‚úÖ Gr√°fica comparativa guardada en: {graph_path}")
    plt.show()

def get_xml_files(directory):
    return {f for f in os.listdir(directory) if f.endswith(".xml")}

def prepare_mixed_config(config_dir, expected_files):
    mixed_dir = tempfile.mkdtemp()
    missing_files = set(expected_files)

    for root, _, files in os.walk(config_dir):
        for file in files:
            if file in expected_files:
                shutil.copy2(os.path.join(root, file), os.path.join(mixed_dir, file))
                missing_files.discard(file)

    if missing_files and os.path.exists(SYSTEM_WRONG):
        print(f"    üîÑ Usando {len(missing_files)} archivos de systemWrong")
        for root, _, files in os.walk(SYSTEM_WRONG):
            for file in files:
                if file in missing_files:
                    shutil.copy2(os.path.join(root, file), os.path.join(mixed_dir, file))

    return mixed_dir

# Main notebook execution
os.makedirs(RESULTS_DIR, exist_ok=True)
test_dir = "test/xml/"

if not os.path.exists(test_dir):
    print(f"‚ùå Error: No existe {test_dir}")
else:
    expected_files = get_xml_files(test_dir)
    if not expected_files:
        print("‚ùå Error: No hay archivos XML en test/xml/")
    else:
        print(f"\nüîç Archivos de referencia: {len(expected_files)} en test/xml/")

        for system_path in SYSTEM_DIRS:
            if not os.path.exists(system_path):
                print(f"\n‚ö†Ô∏è  Sistema no encontrado: {system_path}")
                continue

            print(f"\nüìä Procesando: {system_path}")
            configs = sort_by_prompt_number(os.listdir(system_path))

            df_eval = {}
            for eval_type in EVAL_TYPES:
                print(f"\nüîß Evaluaci√≥n: {eval_type}")
                all_stats = {}

                for config in configs:
                    config_dir = os.path.join(system_path, config)
                    if not os.path.isdir(config_dir):
                        continue

                    print(f"  ‚öôÔ∏è  Config: {config}")
                    mixed_dir = prepare_mixed_config(config_dir, expected_files)

                    try:
                        output = run_evaluation(test_dir, mixed_dir, eval_type)
                        stats = parse_evaluation_output(output)
                        all_stats[config] = stats
                        print(f"    ‚úÖ Leak: {stats['Leak']:.3f} | F1: {stats['F1']:.3f} | P: {stats['Precision']:.3f} | R: {stats['Recall']:.3f}")
                    finally:
                        shutil.rmtree(mixed_dir)

                if all_stats:
                    safe_name = sanitize_filename(system_path)
                    df = pd.DataFrame.from_dict(all_stats, orient="index").reset_index().rename(columns={"index": "Config"})
                    generar_grafica_unica(df, safe_name, eval_type)
                    df_eval[eval_type] = df

            if "ner" in df_eval and "spans" in df_eval:
                generar_grafica_comparativa(df_eval["ner"], df_eval["spans"], safe_name)
```

# Nota sobre la m√©trica Leak Score en el subtrack NER

El **Leak Score** es una m√©trica espec√≠fica del subtrack **NER** porque est√° directamente relacionada con la detecci√≥n de categor√≠as de **PHI** (Informaci√≥n de Identificaci√≥n Personal).

En el caso de **SPANS**, dado que la evaluaci√≥n se centra √∫nicamente en intervalos de texto, esta m√©trica no se considera relevante y, por lo tanto, no se incluye en el script de evaluaci√≥n de **MedDocAN**.
