```python
import os
import subprocess
import re
import pandas as pd
import matplotlib.pyplot as plt
import shutil
import tempfile

# Configuraci√≥n
RESULTS_DIR = "results"
SYSTEM_DIRS = [
    "procesados_meddocan/systemLlama3.3/entidades/",
    "procesados_meddocan/systemLlama3.3/etiquetado/",
    "procesados_meddocan/systemPhi4/entidades/",
    "procesados_meddocan/systemPhi4/etiquetado/"
]
SYSTEM_WRONG = "systemWrong"
EVAL_TYPES = ["ner", "spans"]
GRAPH_FILE = os.path.join(RESULTS_DIR, "metrics__{}__{}.png")

def run_evaluation(test_dir, system_dir, eval_type):
    cmd = ["python", "evaluate.py", "i2b2", eval_type, test_dir, system_dir]
    result = subprocess.run(cmd, capture_output=True, text=True)
    return result.stdout

def sort_by_prompt_number(lst):
    def extract_number(s):
        match = re.search(r'(\d+)', s)
        return int(match.group(1)) if match else float('inf')
    return sorted(lst, key=extract_number)

def sanitize_filename(name):
    return name.strip("/").replace("/", "_")

def parse_evaluation_output(output):
    stats = {"Leak": 0, "Precision": 0, "Recall": 0, "F1": 0}
    for line in output.splitlines():
        if "Leak" in line and not "Leakage" in line:
            leak = re.search(r"Leak\s+([\d.]+)", line)
            if leak: stats["Leak"] = float(leak.group(1))
        elif "Precision" in line:
            prec = re.search(r"Precision\s+([\d.]+)", line)
            if prec: stats["Precision"] = float(prec.group(1))
        elif "Recall" in line:
            rec = re.search(r"Recall\s+([\d.]+)", line)
            if rec: stats["Recall"] = float(rec.group(1))
        elif "F1" in line:
            f1 = re.search(r"F1\s+([\d.]+)", line)
            if f1: stats["F1"] = float(f1.group(1))
    return stats

def generar_grafica_unica(df, system_name, eval_type):
    plt.figure(figsize=(14, 7))
    if eval_type != "spans":
        plt.plot(df["Config"], df["Leak"], label="Leak", marker="o", linestyle="--", color="red")
    plt.plot(df["Config"], df["Precision"], label="Precision", marker="s", color="blue")
    plt.plot(df["Config"], df["Recall"], label="Recall", marker="^", color="green")
    plt.plot(df["Config"], df["F1"], label="F1", marker="D", color="purple")

    plt.title(f"M√©tricas ({eval_type}) - {system_name}", fontsize=14)
    plt.xlabel("Configuraci√≥n", fontsize=12)
    plt.ylabel("Score", fontsize=12)
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()

    graph_path = GRAPH_FILE.format(system_name, eval_type)
    plt.savefig(graph_path)
    print(f"‚úÖ Gr√°fica guardada en: {graph_path}")
    plt.show()

def generar_grafica_comparativa(df_ner, df_spans, system_name):
    plt.figure(figsize=(14, 7))

    for metric in ["Precision", "Recall", "F1"]:
        plt.plot(df_ner["Config"], df_ner[metric], label=f"{metric} (NER)", linestyle="--", marker="o")
        plt.plot(df_spans["Config"], df_spans[metric], label=f"{metric} (Spans)", linestyle="-", marker="s")

    plt.title(f"Comparativa NER vs Spans - {system_name}", fontsize=14)
    plt.xlabel("Configuraci√≥n", fontsize=12)
    plt.ylabel("Score", fontsize=12)
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()

    graph_path = GRAPH_FILE.format(system_name, "comparativa")
    plt.savefig(graph_path)
    print(f"‚úÖ Gr√°fica comparativa guardada en: {graph_path}")
    plt.show()

def get_xml_files(directory):
    return {f for f in os.listdir(directory) if f.endswith(".xml")}

def prepare_mixed_config(config_dir, expected_files):
    mixed_dir = tempfile.mkdtemp()
    missing_files = set(expected_files)

    for root, _, files in os.walk(config_dir):
        for file in files:
            if file in expected_files:
                shutil.copy2(os.path.join(root, file), os.path.join(mixed_dir, file))
                missing_files.discard(file)

    if missing_files and os.path.exists(SYSTEM_WRONG):
        print(f"    üîÑ Usando {len(missing_files)} archivos de systemWrong")
        for root, _, files in os.walk(SYSTEM_WRONG):
            for file in files:
                if file in missing_files:
                    shutil.copy2(os.path.join(root, file), os.path.join(mixed_dir, file))

    return mixed_dir

# Main notebook execution
os.makedirs(RESULTS_DIR, exist_ok=True)
test_dir = "test/xml/"

if not os.path.exists(test_dir):
    print(f"‚ùå Error: No existe {test_dir}")
else:
    expected_files = get_xml_files(test_dir)
    if not expected_files:
        print("‚ùå Error: No hay archivos XML en test/xml/")
    else:
        print(f"\nüîç Archivos de referencia: {len(expected_files)} en test/xml/")

        for system_path in SYSTEM_DIRS:
            if not os.path.exists(system_path):
                print(f"\n‚ö†Ô∏è  Sistema no encontrado: {system_path}")
                continue

            print(f"\nüìä Procesando: {system_path}")
            configs = sort_by_prompt_number(os.listdir(system_path))

            df_eval = {}
            for eval_type in EVAL_TYPES:
                print(f"\nüîß Evaluaci√≥n: {eval_type}")
                all_stats = {}

                for config in configs:
                    config_dir = os.path.join(system_path, config)
                    if not os.path.isdir(config_dir):
                        continue

                    print(f"  ‚öôÔ∏è  Config: {config}")
                    mixed_dir = prepare_mixed_config(config_dir, expected_files)

                    try:
                        output = run_evaluation(test_dir, mixed_dir, eval_type)
                        stats = parse_evaluation_output(output)
                        all_stats[config] = stats
                        print(f"    ‚úÖ Leak: {stats['Leak']:.3f} | F1: {stats['F1']:.3f} | P: {stats['Precision']:.3f} | R: {stats['Recall']:.3f}")
                    finally:
                        shutil.rmtree(mixed_dir)

                if all_stats:
                    safe_name = sanitize_filename(system_path)
                    df = pd.DataFrame.from_dict(all_stats, orient="index").reset_index().rename(columns={"index": "Config"})
                    generar_grafica_unica(df, safe_name, eval_type)
                    df_eval[eval_type] = df

            if "ner" in df_eval and "spans" in df_eval:
                generar_grafica_comparativa(df_eval["ner"], df_eval["spans"], safe_name)
```

# Nota sobre la m√©trica Leak Score en el subtrack NER

El **Leak Score** es una m√©trica espec√≠fica del subtrack **NER** porque est√° directamente relacionada con la detecci√≥n de categor√≠as de **PHI** (Informaci√≥n de Identificaci√≥n Personal).

En el caso de **SPANS**, dado que la evaluaci√≥n se centra √∫nicamente en intervalos de texto, esta m√©trica no se considera relevante y, por lo tanto, no se calcula en el script de evaluaci√≥n de **MEDDOCAN**.

# An√°lisis de la degradaci√≥n del Prompt 5 vs Prompt 4 en m√©tricas (ner) - Llama3.3 en entidades
## An√°lisis comparativo por tags:


```python
from generate_stats import *
from pathlib import Path
GOLD_DIR = "test/xml"
PROMPT_1_DIR = "procesados_meddocan/systemLlama3.3/entidades/prompt4"
PROMPT_2_DIR = "procesados_meddocan/systemLlama3.3/entidades/prompt5"
OUTPUT_DIR = "analisis_comparativo_4_heatmaps"

Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)

print("Comparando prompts...")
analyzer = MeddocanAnalyzer()

print("Analizando Prompt 4...")
metrics_1 = analyzer.analyze_annotations(GOLD_DIR, PROMPT_1_DIR, prompt_name="Prompt_4")

print("\nAnalizando Prompt 5...")
metrics_2 = analyzer.analyze_annotations(GOLD_DIR, PROMPT_2_DIR, prompt_name="Prompt_5")

# Crear DataFrame para comparaci√≥n completa
comparison_data_full = []
# Obtener todos los tags √∫nicos de ambos conjuntos de m√©tricas
all_tags = sorted(list(set(list(metrics_1.keys()) + list(metrics_2.keys()))))

# M√©tricas por defecto si un tag no existe en un prompt
default_metric_values = {'precision': 0.0, 'recall': 0.0, 'f1': 0.0, 'tp': 0, 'fp': 0, 'fn': 0}

for tag in all_tags:
    m1_tag_metrics = metrics_1.get(tag, default_metric_values.copy())
    m2_tag_metrics = metrics_2.get(tag, default_metric_values.copy())

    comparison_data_full.append({
        'Tag': tag,
        'P4_Precision': m1_tag_metrics['precision'],
        'P4_Recall': m1_tag_metrics['recall'],
        'P4_F1': m1_tag_metrics['f1'],
        'P5_Precision': m2_tag_metrics['precision'],
        'P5_Recall': m2_tag_metrics['recall'],
        'P5_F1': m2_tag_metrics['f1'],
        'Diff_Precision': m2_tag_metrics['precision'] - m1_tag_metrics['precision'],
        'Diff_Recall': m2_tag_metrics['recall'] - m1_tag_metrics['recall'],
        'Diff_F1': m2_tag_metrics['f1'] - m1_tag_metrics['f1'],
    })

df_comparison_full = pd.DataFrame(comparison_data_full)
df_comparison_full = df_comparison_full.set_index('Tag')

# Guardar DataFrame completo (√∫til para referencia y otras tablas)
excel_path = Path(OUTPUT_DIR) / "comparacion_completa_metricas_P4_vs_P5.xlsx"
df_comparison_full.reset_index().to_excel(excel_path, index=False)
print(f"\nTabla de comparaci√≥n completa guardada en: {excel_path}")

# --- Generaci√≥n de los 4 Heatmaps Comparativos ---
plt.style.use('seaborn-v0_8-whitegrid')

def plot_comparative_heatmap(df_data, metric_basename, output_dir_path):
    cols_to_plot = [f'P4_{metric_basename}', f'P5_{metric_basename}', f'Diff_{metric_basename}']
    
    # Verificar si todas las columnas necesarias existen
    if not all(col in df_data.columns for col in cols_to_plot):
        print(f"Advertencia: Faltan columnas para el heatmap de {metric_basename}. Columnas requeridas: {cols_to_plot}. Columnas disponibles: {df_data.columns.tolist()}")
        return

    plt.figure(figsize=(10, max(6, len(df_data.index) * 0.5)))
    
    # Usar 'coolwarm' para que el centro (0 para Diff) sea blanco
    # y los valores positivos/negativos tengan colores distintos.
    sns.heatmap(df_data[cols_to_plot],
                annot=True,
                cmap="coolwarm",
                center=0, # Importante para la columna 'Diff'
                fmt=".3f",
                linewidths=.5,
                cbar_kws={'label': f'{metric_basename} Score / Diferencia (P5 - P4)'})
    
    plt.title(f'Comparaci√≥n {metric_basename} (Prompt 4 vs Prompt 5) por Entidad')
    plt.ylabel('Entidad')
    plt.tight_layout()
    
    fig_path = output_dir_path / f"heatmap_comparativo_{metric_basename.lower()}.png"
    plt.savefig(fig_path, dpi=300)
    print(f"Heatmap comparativo de {metric_basename} guardado en: {fig_path}")
    plt.show()

# 1. Heatmap de F1-score
plot_comparative_heatmap(df_comparison_full, 'F1', Path(OUTPUT_DIR))

# 2. Heatmap de Precision
plot_comparative_heatmap(df_comparison_full, 'Precision', Path(OUTPUT_DIR))

# 3. Heatmap de Recall
plot_comparative_heatmap(df_comparison_full, 'Recall', Path(OUTPUT_DIR))

print(f"\nHeatmaps comparativos guardados en el directorio: {OUTPUT_DIR}")
print("An√°lisis comparativo completado.")
```

```python
import os
import matplotlib.pyplot as plt
import numpy as np

# Rutas de las carpetas
folder4 = "systemLlama3.3/prompt4"
folder5 = "systemLlama3.3/prompt5"

# Obtener lista de archivos .txt en ambas carpetas
files4 = set(f for f in os.listdir(folder4) if f.endswith('.txt'))
files5 = set(f for f in os.listdir(folder5) if f.endswith('.txt'))
common_files = files4 & files5  # Archivos presentes en ambas carpetas

# Comparar longitudes
longer_in_4 = 0
longer_in_5 = 0
equal_length = 0
file_differences = []  # Ahora almacenaremos (archivo, diferencia, len4, len5)

for file in common_files:
    with open(os.path.join(folder4, file), 'r', encoding='utf-8') as f4, \
         open(os.path.join(folder5, file), 'r', encoding='utf-8') as f5:
        len4 = len(f4.read())
        len5 = len(f5.read())
        difference = len4 - len5
        
        if len4 > len5:
            longer_in_4 += 1
        elif len5 > len4:
            longer_in_5 += 1
        else:
            equal_length += 1
            
        file_differences.append((file, difference, len4, len5))

total_files = len(common_files)
percentage_4_longer = (longer_in_4 / total_files) * 100
percentage_5_longer = (longer_in_5 / total_files) * 100
percentage_equal = (equal_length / total_files) * 100

# Gr√°fico 1: Porcentaje de veces que prompt4 es m√°s largo que prompt5
labels = ['Prompt4 m√°s largo', 'Prompt5 m√°s largo', 'Igual longitud']
percentages = [percentage_4_longer, percentage_5_longer, percentage_equal]

plt.figure(figsize=(10, 5))
plt.bar(labels, percentages, color=['blue', 'orange', 'green'])
plt.title('Comparaci√≥n de longitud entre Prompt4 y Prompt5')
plt.ylabel('Porcentaje de archivos (%)')
plt.ylim(0, 100)
for i, v in enumerate(percentages):
    plt.text(i, v + 1, f"{v:.1f}%", ha='center')
plt.show()

# Gr√°fico 2: Diferencia de longitud (prompt4 - prompt5) por archivo
length_differences = [diff for (file, diff, len4, len5) in file_differences]
plt.figure(figsize=(12, 6))
plt.scatter(range(len(length_differences)), length_differences, alpha=0.6)
plt.axhline(0, color='red', linestyle='--')
plt.title('Diferencia de longitud por archivo (Prompt4 - Prompt5)')
plt.xlabel('Archivo (√≠ndice)')
plt.ylabel('Diferencia de longitud (caracteres)')
plt.grid(True)
plt.show()

# Estad√≠sticas adicionales
print(f"\nEstad√≠sticas:")
print(f"- Total archivos comparados: {total_files}")
print(f"- Prompt4 m√°s largo: {longer_in_4} veces ({percentage_4_longer:.1f}%)")
print(f"- Prompt5 m√°s largo: {longer_in_5} veces ({percentage_5_longer:.1f}%)")
print(f"- Igual longitud: {equal_length} veces ({percentage_equal:.1f}%)")
if length_differences:
    avg_diff = np.mean(length_differences)
    print(f"- Diferencia promedio: {avg_diff:.1f} caracteres (positivo = prompt4 m√°s largo)")

# Ordenar por diferencia absoluta (los m√°s destacados primero)
file_differences_sorted = sorted(file_differences, key=lambda x: abs(x[1]), reverse=True)

# Mostrar top 10
print("\nTOP 10 ARCHIVOS CON MAYOR DIFERENCIA (prompt4 - prompt5):")
print("-" * 90)
print(f"{'Archivo':<40} | {'Long. prompt4':>12} | {'Long. prompt5':>12} | {'Diferencia':>12} | {'Tendencia':<10}")
print("-" * 90)
for file, diff, len4, len5 in file_differences_sorted[:10]:
    tendency = "prompt4+" if diff > 0 else "prompt5+"
    print(f"{file:<40} | {len4:>12} | {len5:>12} | {diff:>12} | {tendency:<10}")

# Separador para top positivo/negativo
top_prompt4_longer = sorted([fd for fd in file_differences if fd[1] > 0], key=lambda x: x[1], reverse=True)[:5]
top_prompt5_longer = sorted([fd for fd in file_differences if fd[1] < 0], key=lambda x: x[1])[:5]

print("\nTOP 5 ARCHIVOS DONDE PROMPT4 ES MUCHO M√ÅS LARGO:")
for file, diff, len4, len5 in top_prompt4_longer:
    print(f"- {file}: +{diff} caracteres (prompt4={len4} vs prompt5={len5})")

print("\nTOP 5 ARCHIVOS DONDE PROMPT5 ES MUCHO M√ÅS LARGO:")
for file, diff, len4, len5 in top_prompt5_longer:
    print(f"- {file}: {diff} caracteres (prompt5 es {abs(diff)} m√°s largo)")
```
