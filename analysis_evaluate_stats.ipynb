```python
from stats import EvaluationSystem, quick_evaluation

# Configuraci√≥n
SYSTEM_DIRS = [
    "procesados_meddocan/systemLlama3.3/entidades/",
    "procesados_meddocan/systemLlama3.3/etiquetado/"
]

EVAL_TYPES = ["ner", "spans"]

# Opci√≥n 1: Uso r√°pido con funci√≥n de conveniencia
print("=== EVALUACI√ìN R√ÅPIDA ===")
results = quick_evaluation(SYSTEM_DIRS, eval_types=EVAL_TYPES)

# Opci√≥n 2: Uso avanzado con control completo
print("\n=== EVALUACI√ìN AVANZADA ===")
evaluator = EvaluationSystem(results_dir="results", system_wrong="systemWrong")

# Evaluar sistemas individuales si necesitas m√°s control
for system_path in SYSTEM_DIRS:
    system_results = evaluator.evaluate_system(
        system_path=system_path,
        test_dir="test/xml/",
        eval_types=EVAL_TYPES
    )
    
    # Generar comparativa si hay ambos tipos
    if "ner" in system_results and "spans" in system_results:
        safe_name = evaluator.sanitize_filename(system_path)
        evaluator.generate_comparison_graph(
            system_results["ner"], 
            system_results["spans"], 
            safe_name
        )

print("\n‚úÖ Evaluaci√≥n completada")
```
